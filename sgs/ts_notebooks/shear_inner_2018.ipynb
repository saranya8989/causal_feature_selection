```python
#nio tracks use latitude longitude instead of lat lon
track=pd.read_csv(p1+'2016/namtheun_wp.csv',delimiter=r",")
lon1=track['lon'].to_numpy()
lat1=track['lat'].to_numpy()
pos = arr = np.stack((lat1, lon1), axis=1)
pos.shape
```

```python
#for North Atlantic
lonx=np.mod(lon1,360)
pos = arr = np.stack((lat1, lonx), axis=1)
pos.shape
```

```python
vname0=[ "in_vor200","in_vor600","in_vor850","in_hum200","in_hum600","in_hum850"]
```

```python
tc_irad=np.empty((pos.shape[0],4))
tc_irad[:,0] = pos[:,0]-2
tc_irad[:,1] = pos[:,0]+2
tc_irad[:,2] = pos[:,1]-2
tc_irad[:,3] = pos[:,1]+2
```

```python
tc_irad
```

```python
dv200=np.zeros((pos.shape[0],))
dv600=np.zeros((pos.shape[0],))
dv850=np.zeros((pos.shape[0],))
dh200=np.zeros((pos.shape[0],))
dh600=np.zeros((pos.shape[0],))
dh850=np.zeros((pos.shape[0],))
```

```python
dm02=dm2.var138[1942:1989,7]
dm03=dm2.var138[1942:1989,12]
dm04=dm2.var138[1942:1989,15]
dm05=dm3.var157[1942:1989,7]
dm06=dm3.var157[1942:1989,12]
dm07=dm3.var157[1942:1989,15]
```

```python
for it in range(tc_irad.shape[0]):
    latn, lats, lone, lonw = tc_irad[it,:]
    dv200[it]=dm02[it,:,:].sel(lat=slice(lats,latn),lon=slice(lone,lonw)).mean(dim=['lat','lon'])
    dv600[it]=dm03[it,:,:].sel(lat=slice(lats,latn),lon=slice(lone,lonw)).mean(dim=['lat','lon'])
    dv850[it]=dm04[it,:,:].sel(lat=slice(lats,latn),lon=slice(lone,lonw)).mean(dim=['lat','lon'])
    dh200[it]=dm05[it,:,:].sel(lat=slice(lats,latn),lon=slice(lone,lonw)).mean(dim=['lat','lon'])
    dh600[it]=dm06[it,:,:].sel(lat=slice(lats,latn),lon=slice(lone,lonw)).mean(dim=['lat','lon'])
    dh850[it]=dm07[it,:,:].sel(lat=slice(lats,latn),lon=slice(lone,lonw)).mean(dim=['lat','lon'])
```

```python
dv200
```

```python

df1 = pd.DataFrame(data=dv200, columns=["in_vor200"])
df2 = pd.DataFrame(data=dv600, columns=["in_vor600"])
df3 = pd.DataFrame(data=dv850, columns=["in_vor850"])
df4 = pd.DataFrame(data=dh200, columns=["in_hum200"])
df5 = pd.DataFrame(data=dh600, columns=["in_hum600"])
df6 = pd.DataFrame(data=dh850, columns=["in_hum850"])
```

```python
df=pd.concat([df1,df2,df3,df4,df5,df6], axis=1, join='inner')
df
```

```python
df.to_csv(target+'2016_inner_wp_namtheun.csv')
```
