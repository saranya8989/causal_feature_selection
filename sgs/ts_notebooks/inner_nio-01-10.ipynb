```python
 track=pd.read_csv(p1+'nio/nio_20051204.csv',delimiter=r",")
```

```python
vname0=["div100","div200","div250","div300","div400","div500","eqt1000","eqt200","eqt250","eqt300",
        "eqt400","eqt500","eqt600","eqt700","eqt850","eqt925","sh850_200","sh850_250","sh925_200","sh925_250","ws10","wstrss10"]
```

```python
lat1=track['latitude'].to_numpy()
lon1=track['longitude'].to_numpy()
pos = arr = np.stack((lat1, lon1), axis=1)
pos.shape[0]
```

```python
tc_irad=np.empty((pos.shape[0],4))
tc_irad[:,0] = pos[:,0]-2
tc_irad[:,1] = pos[:,0]+2
tc_irad[:,2] = pos[:,1]-2
tc_irad[:,3] = pos[:,1]+2
```

```python
df0={}
for v in vname0:
    df0[v]=np.zeros((tc_irad.shape[0],))
```

```python
for it in range(tc_irad.shape[0]):
    latn, lats, lone, lonw = tc_irad[it,:]
    for iv,v in enumerate(vname0):
        df0[v][it]=dm1[v][it,:,:].sel(lat=slice(latn,lats),lon=slice(lone,lonw)).mean(dim=['lat','lon'])
```

```python
dft=pd.DataFrame.from_dict(df0)
```
