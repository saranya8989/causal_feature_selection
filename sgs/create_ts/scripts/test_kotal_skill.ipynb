```python
crps_ensemble = xs.crps_ensemble(tc_kgpp, x, dim=[])
print(crps_ensemble)
```

```python
crps_ensemble.mean('time').plot()
```

```python
threshold_brier_score = xs.threshold_brier_score(tc_kgpp, x, 0.5, dim=None)
print(threshold_brier_score)
```

```python
rel = xs.reliability(tc_kgpp > 0.5, (x > 0.5).mean("member"))
print(rel)
```

```python
# ROC for probabilistic forecasts and bin_edges='continuous' default
roc = xs.roc(tc_kgpp > 0.0, (x > 0.0).mean("member"), return_results='all_as_metric_dim')
#roc2 = xs.roc(tc_igpp > 0.0, (y > 0.0).mean("member"), return_results='all_as_metric_dim')

plt.figure(figsize=(10, 8))
plt.plot([0, 1], [0, 1], 'k:')
roc.to_dataset(dim='metric').plot.scatter(y='true positive rate', x='false positive rate')
roc.sel(metric='area under curve').values[0]

```
