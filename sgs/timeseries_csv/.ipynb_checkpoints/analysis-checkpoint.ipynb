```python
# Imports
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline     
## use `%matplotlib notebook` for interactive figures
# plt.style.use('ggplot')
import sklearn

import tigramite
from tigramite import data_processing as pp
from tigramite.toymodels import structural_causal_processes as toys

from tigramite import plotting as tp
from tigramite.pcmci import PCMCI
from tigramite.independence_tests import ParCorr, GPDC, CMIknn, CMIsymb

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("time_series_bg_novmax.csv",sep='\s+')
df.head()
```

```python
df = pd.read_csv("time_series_bg_novmax.csv", sep='\s+',
                 parse_dates=[0])
```

```python
df.head()
```

```python
storm_np=df.values
```

```python
storm_np
```

```python
#lowe level wind speeds
raw_wind=np.empty((55,4))
raw_wind[:,0] = storm_np[:,1]
raw_wind[:,1] = storm_np[:,2]
raw_wind[:,2] = storm_np[:,3]
raw_wind[:,3] = storm_np[:,4]
```

```python


# Initialize dataframe object, specify time axis and variable names
#var_names = ["new_confirmed", "new_deceased","total_confirmed","total_deceased"]
var_names = [r"w_10m", r"w_850m",r'w_700hPa',r'w_600hpa']
dataframe = pp.DataFrame(raw_wind, 
                         datatime = np.arange(len(raw_wind)), 
                         var_names=var_names)
```

```python
tp.plot_timeseries(dataframe); plt.show()
```

```python
parcorr = ParCorr(significance='analytic') #partial correlation
pcmci = PCMCI(
    dataframe=dataframe, 
    cond_ind_test=parcorr,
    verbosity=1)
```

```python
correlations = pcmci.get_lagged_dependencies(tau_max=20, val_only=True)['val_matrix']
lag_func_matrix = tp.plot_lagfuncs(val_matrix=correlations, setup_args={'var_names':var_names, 
                                    'x_base':5, 'y_base':.5}); plt.show()
```

```python
pcmci.verbosity = 1
results = pcmci.run_pcmci(tau_max=1, pc_alpha=None, alpha_level=0.01)
```

```python
print("p-values")
print (results['p_matrix'].round(3))
print("MCI partial correlations")
print (results['val_matrix'].round(2))
```

```python
q_matrix = pcmci.get_corrected_pvalues(p_matrix=results['p_matrix'], tau_max=1, fdr_method='fdr_bh')
pcmci.print_significant_links(
        p_matrix = q_matrix,
        val_matrix = results['val_matrix'],
        alpha_level = 0.01)
graph = pcmci.get_graph_from_pmatrix(p_matrix=q_matrix, alpha_level=0.01, 
            tau_min=0, tau_max=1, selected_links=None)
results['graph'] = graph
```

```python
tp.plot_graph(
    val_matrix=results['val_matrix'],
    graph=results['graph'],
    var_names=var_names,
    link_colorbar_label='cross-MCI',
    node_colorbar_label='auto-MCI',
    ); plt.show()
```
